```python
sns.lmplot(data=activity, x='TotalSteps', y='Calories', line_kws={"color": "red"})
plt.title('Steps vs Calories with Trend Line')
plt.show()
```

```python
asleep_minutes = sleep[sleep['value'] == 1].groupby('Id').size().reset_index(name='TotalMinutesAsleep')
```

```python
minutes_in_bed = sleep.groupby('Id').size().reset_index(name='TotalMinutesInBed')
```

```python
sleep_summary = pd.merge(asleep_minutes, minutes_in_bed, on='Id')
```

```python
sns.lmplot(data=sleep_summary,x='TotalMinutesInBed',y='TotalMinutesAsleep',height=6,aspect=1.5,scatter_kws={'color': 'skyblue', 'alpha': 0.7},line_kws={'color': 'red'})

plt.title('Scatter Plot: Minutes Asleep vs Minutes in Bed')
plt.xlabel('Total Minutes in Bed')
plt.ylabel('Total Minutes Asleep')

plt.tight_layout()
plt.show()
```
